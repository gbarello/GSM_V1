```python
import numpy as np

F = np.loadtxt("./image_processing/kernels.csv")
F = np.reshape(F,[F.shape[0],np.int32(np.sqrt(F.shape[1])),np.int32(np.sqrt(F.shape[1]))])
print(F.shape)
```

```python
from matplotlib import pyplot as plt
```

```python
fig,subs = plt.subplots(6,10)

for k in range(6):
    for j in range(10):
        subs[k][j].matshow(F[k + j * 6])
plt.show()
```
